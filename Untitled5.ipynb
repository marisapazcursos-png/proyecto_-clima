```python
import requests
import pandas as pd
import matplotlib.pyplot as plt
from datetime import datetime

# -------------------------------------------------------------------------
# CONFIGURACIÓN Y CREDENCIALES
# -------------------------------------------------------------------------

# Tal como solicitaste, pedimos la API Key por pantalla
api_key = input(" Introduce tu API Key de AEMET: ")

# URL para obtener todas las observaciones convencionales
url_aemet = "https://opendata.aemet.es/opendata/api/observacion/convencional/todas"
headers = {'cache-control': "no-cache", 'api_key': api_key}

print("\n🤖 Robot: Iniciando conexión con AEMET...")

# -------------------------------------------------------------------------
# VIAJE 1: Pedir permiso y dirección secreta
# -------------------------------------------------------------------------
respuesta_1 = requests.get(url_aemet, headers=headers)
datos_permiso = respuesta_1.json()

if datos_permiso['estado'] == 200:
    url_secreta = datos_permiso['datos']
    print(f"✅ Robot: Permiso concedido. Descargando datos reales...")

    # ---------------------------------------------------------------------
    # VIAJE 2: Recoger el "paquete" de datos climáticos
    # ---------------------------------------------------------------------
    respuesta_final = requests.get(url_secreta)
    datos_climaticos = respuesta_final.json()

    # Convertimos la lista de datos en una Tabla (DataFrame) [cite: 1313]
    tabla_completa = pd.DataFrame(datos_climaticos)

    # ---------------------------------------------------------------------
    # FILTRADO: Centrarse en Tenerife [cite: 1284, 1289]
    # ---------------------------------------------------------------------
    # Filtramos la columna 'ubi' (ubicación) para que contenga "TENERIFE"
    tabla_tenerife = tabla_completa[tabla_completa['ubi'].str.contains("TENERIFE", na=False)].copy()

    if tabla_tenerife.empty:
        print("⚠️ Robot: No se han encontrado estaciones con el nombre 'TENERIFE'. Mostrando datos generales.")
        tabla_clima = tabla_completa.head(15).copy()
    else:
        tabla_clima = tabla_tenerife
        print(f"🏝️ Robot: He filtrado {len(tabla_clima)} estaciones de Tenerife.")

    # ---------------------------------------------------------------------
    # ANÁLISIS: Máximas, Mínimas y Alertas [cite: 1406, 1407, 1412]
    # ---------------------------------------------------------------------
    max_temp = tabla_clima['ta'].max()
    min_temp = tabla_clima['ta'].min()

    ciudad_calor = tabla_clima[tabla_clima['ta'] == max_temp]['ubi'].values[0]
    ciudad_frio = tabla_clima[tabla_clima['ta'] == min_temp]['ubi'].values[0]

    print(f"\n📊 RESUMEN ACTUAL:")
    print(f"🔥 Máxima: {max_temp}°C en {ciudad_calor}")
    print(f"❄️ Mínima: {min_temp}°C en {ciudad_frio}")

    # ---------------------------------------------------------------------
    # VISUALIZACIÓN: Gráfico de Barras con Extremos [cite: 924]
    # ---------------------------------------------------------------------
    colores = []
    for t in tabla_clima['ta']:
        if t == max_temp: colores.append('red')    # Rojo para la más calurosa
        elif t == min_temp: colores.append('blue') # Azul para la más fría
        else: colores.append('skyblue')

    plt.figure(figsize=(12, 6))
    barras = plt.bar(tabla_clima['ubi'], tabla_clima['ta'], color=colores)

    plt.title(f'Temperaturas en Tenerife - {datetime.now().strftime("%d/%m/%Y")}', fontsize=15)
    plt.xticks(rotation=45, ha='right')
    plt.ylabel('Temperatura (°C)')

    # Añadir el número exacto sobre cada barra [cite: 924]
    for barra in barras:
        yval = barra.get_height()
        plt.text(barra.get_x() + barra.get_width()/2, yval + 0.1, yval, ha='center', va='bottom')

    plt.tight_layout()
    plt.show()

    # ---------------------------------------------------------------------
    # GUARDADO: Histórico automático [cite: 1385, 1386]
    # ---------------------------------------------------------------------
    fecha_hoy = datetime.now().strftime("%Y-%m-%d")
    nombre_archivo = f"clima_tenerife_{fecha_hoy}.csv"
    tabla_clima.to_csv(nombre_archivo, index=False)
    print(f"\n💾 Robot: Misión cumplida. Datos guardados en '{nombre_archivo}'.")

else:
    print(f"❌ Error: AEMET dice: {datos_permiso['descripcion']}")
```
